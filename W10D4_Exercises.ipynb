```python
# Qual è la media della pressione sanguigna diastolica per le diverse fasce di età?
```

```python
db["Fascia Età"] = pd.cut(db["Age (years)"],
                          bins=[0, 29, 39, 49, 59, 69, 120],
                          labels=["20-29", "30-39", "40-49", "50-59", "60-69", "70+"])

# Calcola media pressione per fascia
media_pressione = db.groupby("Fascia Età",observed=True)["Diastolic blood pressure (mm Hg)"].mean().round(2)
print(media_pressione)
```

```python
#Essendo i valori di bmi numerosi, creo delle fascie di bmi
insurance["fascia bmi"] = pd.cut(insurance["bmi"],
                          bins=[0, 20, 29, 39, 49, 59],
                          labels=["<20", "20-29", "30-39", "40-49", ">50"])
insurance
```

```python
# analiziamo la media delle spese mediche per fascia di bmi
insurance.groupby("fascia bmi", observed=True)["charges"].mean().round(1)
```

```python
# Possiamo notare un aumento di spese mediche legate all'aumento del bmi index
```

```python
#voglio vedere quante persone ci sono in ogni fascia di bmi
insurance.groupby("fascia bmi", observed= True)["fascia bmi"].count()
```

```python
#Quali sono i descrittori statistici di bmi?
```

```python
describe_bmi = insurance["bmi"].describe().round(2)
describe_bmi
```

```python
#Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?
```

```python
#calcolo di ogni singolo quartile (25,50,75)
q1 = insurance["bmi"].quantile(0.25)
q2 = insurance["bmi"].quantile(0.50)
q3 = insurance["bmi"].quantile(0.75)
```

```python
#Creo una funzione che assegna il quartile di appartenenza ad ogni bmi di insurance
def assegna_quartile(bmi):
    if bmi <= q1:
        return "Q1"
    elif bmi <= q2:
        return "Q2"
    elif bmi <= q3:
        return "Q3"
    else:
        return "Q4"

#creo la colonna "bmi_quartile" applicando la funzione assegna_quartile
insurance["bmi_quartile"] = insurance["bmi"].apply(assegna_quartile)
```

```python
#Infine ragggruppa per la colonna bmi_quartile calcolando per ogni riga il minimo, la media e il massimo di charges
insurance.groupby("bmi_quartile")["charges"].agg(["min", "mean", "max"]).round(2)
```
