```python
# Importo il file wine.csv
path = "datasets/beginner_datasets/wine.csv"
wine = pd.read_csv(path)
wine
```

```python
# Qual è la media di concentrazione alcolica per ogni qualità? Ci sono differenze? E rispetto alla media totale?
```

```python
wine_quality = pd.DataFrame()
wine_quality["alcohol_mean_per_quality"] = wine.groupby("quality")["alcohol"].mean().round(1)
wine_quality["alcohol_mean"] = wine["alcohol"].mean().round(1)
wine_quality["Difference"] = wine_quality["alcohol_mean_per_quality"] - wine_quality["alcohol_mean"]
wine_quality
```

```python
# da questa analisi posso notare nelle differenze nelle diverse qualità del vino.
# in particolare maggiore è la qualità del vino e maggiore è la sua concentrazione alcolica
# è interessante notare come fino alla qualità 6 la differenza con la media generale sia trascurabile
# ma non appena arriviamo a qualità 7 ed oltre abbiamo un impennata significativa 
```

```python
# C'è differenza nella concentrazione alcolica per vini bianchi e vini rossi?
```

```python
wine.groupby("type")["alcohol"].mean().round(1)
```

```python
# dall'analisi emerge che non ci sono differenze significative tra il type bianco e rosso 
```

```python
# Rifacendo le analisi dei due punti precedenti ma per il pH, cambia qualcosa?
```

```python
wine_ph = pd.DataFrame()
wine_ph["pH_mean_per_quality"] = wine.groupby("quality")["pH"].mean().round(1)
wine_ph["pH_mean"] = wine["pH"].mean().round(1)
wine_ph["difference"] = wine_ph["pH_mean_per_quality"] - wine_ph["pH_mean"]
wine_ph
```

```python
# dall'analisi del pH rispetto alle diverse qualità non ermegono differenze "parlanti" 
```

```python
# E per i solfati?
```

```python
wine_sulphates = pd.DataFrame()
wine_sulphates["sulphates_mean_per_quality"] = wine.groupby("quality")["sulphates"].mean().round(1)
wine_sulphates["sulphates_mean"] = wine["sulphates"].mean().round(1)
wine_sulphates["difference"] = wine_sulphates["sulphates_mean_per_quality"] - wine_sulphates["sulphates_mean"]
wine_sulphates
```

```python
# neanche per i solfati ci sono differenze tra le diverse qualità
```

```python
#Si nota una correlazione tra il tasso di criminalità e il valore delle abitazioni? Come si può spiegare il risultato?
```

```python
#mi faccio descrivere la colonna "medv" per avere visione dei dati presenti
boston["medv"].describe().round(1)
```

```python
# per un analisi più efficace divido in fasce di prezzo la colonna medv
# 5 - 17.0: Bassa
# 17.1 - 21.2: Media-bassa
# 21.3 - 25: Media-alta
# 21.4 a 50 Alta
bins = [0, 17.0, 21.2, 25.0, boston["medv"].max()]
labels = ["Bassa", "Media-bassa", "Media-alta", "Alta"]

#creo la colonna "fascia_prezzo"
boston["fascia_prezzo"] = pd.cut(boston["medv"], bins=bins, labels=labels, include_lowest=True)

# raggruppe per fasce di prezzo e faccio la media della colonna crim in relazione 
boston.groupby("fascia_prezzo", observed=True)["crim"].mean()
```

```python
#Dall'analisi emerge una correlazione tra il valore delle case ed il tasso di criminalità
#In particolare la fascia basse racchiude quasi la totalità della criminialità.
#Evidentemente una casa con unn valore minore è indice di povertà dovuta probabilmente a una disoccupazione maggiore.
# e quindi, in definitiva ad un impennata del tasso di criminalità
```

```python
#Qual è la media del numero di stanze rispetto al rapporto alunni-insegnanti?
```

```python
boston.groupby("ptratio")["rm"].mean().round(1)
```

```python
#Posso notare un inversa proporzione tra il rapporto alunni-insegnanti (ptratio) e il numero medio di stanze per abitazione
#quindi in altre parole maggiore è ptratio e minore è la media di stanze per casa
```

```python
boston.groupby("fascia_prezzo", observed=True)["ptratio"].mean().round(1)
```

```python
boston.groupby("fascia_prezzo", observed=True)["rm"].mean().round(1)
```

```python
#Analizziamo la % di popolazione di condizione più bassa (lstat) con le fasce di prezzo delle case
boston.groupby("fascia_prezzo", observed=True)["lstat"].mean().round(1)
```

```python
#Noto come la ad aumentare del valore delle case diminuiscono le persone in basse condizioni
# significa che la maggioranza delle persone con una bassa condizione è concentrata dove le case hanno un valore più basso e, dall'analisi precedente,
# c'è un tasso di crimininalità più alto
```

```python
# I dati appena analizzati mi dicono che più la fascia di prezzo delle case è alta e minore è il rapporto alunni ed insegnanti 

# C'è una diretta proporzione tra la fascia di prezzo ed il numero medio di stanze per casa

# Quindi arrivo alla conclusione che dove c'è maggiore povertà ci sono meno insegnanti, indice di un scarsa educazione e livello scolastico. 
# Tutto questo è collegabile alle analisi fatte in precedenza sul dataframe Boston.csv
# sarebbe interessante capire se è nato prima l'uovo o la gallina, nel senso: capire se siccome c'è povertà allora c'è una inferiore educazione 
# (magari le insegnanti scappano e/o non c'è l'ambizione di diventare educatori)
# oppure siccome c'è meno educazione allora c'è più povertà
```

```python
employees_df = pd.DataFrame({ 
    'employee_id': [101, 102, 103, 104, 105], 
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'], 
    'department_id': [1, 2, 1, 2, 3] 
})

departments_df = pd.DataFrame({ 
    'department_id': [1, 2, 3], 
    'department_name': ['HR', 'IT', 'Finance'], 
    'location': ['New York', 'San Francisco', 'Chicago'] 
})
```

```python
#Unire questi DataFrame in base alla colonna comune department_id
```

```python
employees_df.merge(departments_df, how = 'inner', on= 'department_id')
```

```python
#Per ogni DataFrame, trasformare la colonna department_id nell'indice
#poi unire i due dataset mediante il metodo .join()
```

```python
employees_df.set_index("department_id", inplace=True)
departments_df.set_index('department_id', inplace=True)
```

```python
employees_df.join(departments_df, on='department_id')
```

```python
# Ci sono differenze nel risultato? Quali? Perché?
```

```python
#Sono due risultati diversi. .Merge è più complesso ma più flessibile.
#Di fatti per utilizzare .join c'è bisogno di settare una colonna come indice in quanto fa l'unione delle tabelle esclusivamente su quelle.
#Queste sono delle limitazioni grandi ma in altri casi è più rapido 
# Questo è quello che ho notato nell'utilizzo. Per quanot riguardo il mero output sono diversi, da una parte .join è più pulito con meno colonne
# ma si nota la veritcalità sull'indice da parte di .join()
```
